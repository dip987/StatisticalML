# Data Pre-processing

In this section, we will go over each column individually and look into what sort of pre-processing is used and an insight into why

## Dropping N/A Rows

To start off, we will drop each row where the target variable is unavaible as these rows cannot be used in training/testing. 

```python
df.dropna(axis=0, subset=['IMDb Votes'], inplace=True)
count = df.count()
plt.bar(count.keys(), count.values)
_ = plt.xticks(rotation='vertical')
```

## Categorical Data
The dataset contains a lot of categorical data such as Genre, Language, County Availability and so on. There are however two issues. 
1. For each entry, there might be multiple entries in the same category i.e. the categories might be non-binary. For example, a movie might have more than one Genre
2. For a lot of the the entries one or more categories are missing. But the missing data is reliant on the other observable data of that row(MAR missing mechanism). For example, which languages a certain show is available in might be unknown but looking at the prodcution house, director, country availability etc., we can make an educated guess

To handle these cases, the categorical data will be vectorized and represented as a sparse matrix. On top of that we will be using different variants of Multi-variate Imputation to fill up the missing data.

## Looking at Actors

The list of actors for any given show gives us a better look at the sparsity of the vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

def custom_tokenizer(text):
    split_text = re.split(r",\s?", text)
    return split_text

actor_data = df['Actors']
actor_data.fillna('others', inplace=True)
actor_vect = CountVectorizer(strip_accents='ascii', tokenizer=custom_tokenizer, lowercase=True)
actor_fitted = actor_vect.fit_transform(actor_data.values)

print(f'The shape of the vectorized data is {actor_fitted.shape}')
```

```python
plt.spy(actor_fitted, markersize=0.5)
```

This is an extremely large matrix with more columns than rows. Note that this is for only a single feature. When stacked together with the rest of the categorical features, the combined feature matrix will be extremely hard to train. Therefore, we need a process to reduce the dimensionality of this featureset.

Viable options that we would like to try out
1. Keep only the top-n number of highest frequency occurences and group the rest into the "others" categorgy
2. Instead of the current encoding scheme, use Mean Encoding related to the target column
3. Use Weight of Evidence(WoE) Encoding


We get similar results when looking at Tags, Directors, Country Availability and Writer

## Relation with Release Date

Plotting the resultant variable w.r.t. will give us a good idea about the trends and consequently how to handle the datetime data associated with our prediction

```python
import seaborn as sns
import datetime

date_data = df[['Release Date', 'IMDb Votes']].copy()
date_data.dropna(inplace=True)
date_data['Release Date'] = date_data['Release Date'].map(lambda string: datetime.datetime.strptime(string, "%d %b %Y"))
# Taking the mean for all shows released on the same day
date_data = date_data.groupby(by=['Release Date']).mean()
# Taking a rolling mean over 7 days
date_data= date_data.rolling(7).mean()
sns.lineplot(data=date_data)
```

Since we are more interested in predicting the reach for shows in the future, zooming a bit more into the more recent values

```python
lower_limit = datetime.datetime.strptime('1 May 2019', "%d %b %Y")
upper_limit = datetime.datetime.strptime('1 Jan 2020', "%d %b %Y")

sns.lineplot(data=date_data)
plt.xlim([lower_limit, upper_limit])
plt.ylim([0., 50000])
```

```python
lower_limit = datetime.datetime.strptime('1 Jan 2020', "%d %b %Y")
upper_limit = datetime.datetime.strptime('1 Jun 2020', "%d %b %Y")

sns.lineplot(data=date_data)
plt.xlim([lower_limit, upper_limit])
plt.ylim([0., 50000])
```

```python
lower_limit = datetime.datetime.strptime('1 Jul 2020', "%d %b %Y")
upper_limit = datetime.datetime.strptime('1 Dec 2020', "%d %b %Y")

sns.lineplot(data=date_data)
plt.xlim([lower_limit, upper_limit])
plt.ylim([0., 50000])
```

Looking at the data at different time windows, we see that there is some a relation between the day of the month and the reach. It seems to peak near the end/beginning of every month. The peak value however depends on the month and year. 
To simplify our model, we would like to use the day of month, month number and the year as features with the year shifted w.r.t to 1980(Where the majority of our data points begin)
